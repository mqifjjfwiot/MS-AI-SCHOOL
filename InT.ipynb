```python
select count(distinct L.characterid) as '총 캐릭터 수', 
count(case when levelaft = 5 then 1 end) as '5 달성 캐릭터 수', 
concat(round(count(case when levelaft = 5 then 1 end) / count(distinct L.characterid) * 100, 2), '%') as '5레벨 도달율',
concat(round(avg(L20.L20H), 3), 'h') as '5레벨 달성 평균 소요시간(hour)', 
concat(round(avg(L20.L20H)/24, 3), 'd') as '5레벨 달성 평균 소요일수(day)'
from levelup as L
inner join new_schema.join as J on J.characterid = L.characterid
left join (
	select L.characterid, J.joindate, L.createdate, timestampdiff(hour, J.joindate, L.createdate) as L20H
	from levelup as L
	join new_schema.join as J on J.characterid = L.characterid
	where L.levelaft = 5 and J.joindate < L.createdate
) as L20 on L20.characterid = L.characterid
```

```python
from pandas.core.arrays.datetimelike import timedelta
time_list = []
for i in range(40) : 
  df_lj2 = df_lj[df_lj['levelaft'] == i].copy()
  time_diff_sum = timedelta(0)
  if len(df_lj2) == 0 :
    pass
  else : 
    for j in range(len(df_lj2)) : 
      time_diff_sum += df_lj2.iloc[j]['createdate'] - df_lj2.iloc[j]['joindate']
    time_list.append((time_diff_sum/len(df_lj2)))

time_list
```
